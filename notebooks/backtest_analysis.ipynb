# AlphaCouncil — Backtest Analysis

Interactive analysis of backtest results, agent performance, and strategy comparison.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sqlalchemy import create_engine, text

engine = create_engine('sqlite:///data/alphacouncil.db')
print('Connected to AlphaCouncil database')
```

## Agent Portfolio Comparison

```python
# Load portfolio snapshots
try:
    df = pd.read_sql('SELECT * FROM agent_portfolio_snapshots', engine)
    print(f'Loaded {len(df)} snapshots')
    display(df.tail(20))
except Exception as e:
    print(f'No data yet: {e}')
```

## Trade Analysis
